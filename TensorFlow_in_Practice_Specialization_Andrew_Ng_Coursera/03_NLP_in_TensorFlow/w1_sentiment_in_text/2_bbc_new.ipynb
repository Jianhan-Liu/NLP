```python
def get_rid_of_stopwords(sentence):
    for word in stopwords:
        word = ' ' + word + ' '
        sentence = sentence.replace(word, ' ')
        sentence = sentence.replace('  ', ' ')
    return sentence
```

```python
text_new = data.text.apply(get_rid_of_stopwords)
```
