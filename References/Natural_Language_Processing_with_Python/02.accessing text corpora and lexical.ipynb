```python
def get_analysis(text):
    for fileid in text.fileids():
        num_chars = len(text.raw(fileid))  # 总标识符数，含标点符号
        num_words = len(text.words(fileid))  # 总词汇数
        num_sents = len(text.sents(fileid))  # 总句子数
        num_vocab = len(set([w.lower() for w in text.words(fileid)]))  # 总词类数
        print(f"{int(num_chars/num_words)}, {int(num_words/num_sents)}, {int(num_words/num_vocab)}, {fileid}")
```

```python
get_analysis(gutenberg)
```

```python
from nltk.corpus import brown
```

```python
brown.categories()
```

```python
brown.words(categories='news')
```

```python
brown.sents(categories=['news', 'editorial', 'reviews'])
```

```python
genres = ['news','religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd = nltk.ConditionalFreqDist((genres, word) for genres in brown.categories() for word in brown.words(categories=genres))
cfd.tabulate(condtions=genres, samples=modals)
```

```python
from nltk.corpus import inaugural
```

```python
inaugural.fileids()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
cfd = nltk.ConditionalFreqDist(
    (target, file[:4])
    for file in inaugural.fileids()
    for target in ['america', 'citizen']
    for w in inaugural.words(file) if w.lower().startswith(target)
)
cfd.plot()
```

```python
from nltk.corpus import udhr
```

```python
languages = [x.split('-')[0] for x in udhr.fileids()]
languages_ = ['Chickasaw', 'English', 'German_Deutsch', 'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
udhr.words()
```

```python
cfd = nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages_
    for word in udhr.words(lang+'-Latin1')
)
plt.figure(figsize=(10,10))
cfd.plot(cumulative=1)
```
